# Preview nutrition data

## Load in daters

```python
import pandas as pd
foods = pd.read_csv('data/food.csv').set_index('fdc_id')
food_nutrients = pd.read_csv('data/food_nutrient.csv').set_index(['fdc_id','nutrient_id'])
nutrients = pd.read_csv('data/nutrients.csv').set_index('id')
```

## Find raw broccoli

```python
raw_broccoli = foods[foods['description'] == 'Broccoli, raw']
raw_broccoli
```

## Load broccoli nutrients

```python
broccoli_nutrient_ids = food_nutrients.loc[raw_broccoli.index[0], :]
broccoli_nutrient_ids
```

## Load nutrient names

```python
broccoli_nutrients = nutrients.loc[broccoli_nutrient_ids.index]
broccoli_nutrients
```

## Find KCal row

```python
nutrient_kcal = broccoli_nutrients[broccoli_nutrients['name'] == 'Energy']
nutrient_kcal
```

## How many KCals in broccoli?

```python
broccoli_kcals = broccoli_nutrient_ids.loc[nutrient_kcal.index]
str(broccoli_kcals.iloc[0]['amount']) + ' ' + nutrient_kcal.iloc[0]['unit_name']
```
